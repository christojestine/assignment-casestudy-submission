# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing Excel data set

```python
data=pd.read_excel(r"D:\ICT Academy\Assignment\iris.xls")
```

# Displaying the Columns of the Data set

```python
data.head()
```

# Calculating mean of each column

```python
meansl=data['SL'].mean()
meansw=data['SW'].mean()
meanpl=data['PL'].mean()
meanpw=data['PW'].mean()
```

```python
print("The mean of first column SL is",meansl)
print("The mean of second column SW is",meansw)
print("The mean of third column PL is",meanpl)
print("The mean of fourth column PW is",meanpw)
```

# Data visualizations


```python
data['Classification'].unique()
```

```python
data1=data[data["Classification"]=="Iris-setosa"]
data2=data[data["Classification"]=="Iris-versicolor"]
data3=data[data["Classification"]=="Iris-virginica"]
```

```python
data1.head(2)
```

```python
data2.head(2)
```

```python
data3.head(2)
```

```python
plt.style.use('seaborn')

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.plot(data1["SL"],c='k')
plt.plot(data1["PL"],c='b')
plt.legend(['PL','SL'])
plt.title("Relation between Sepal Lenght and Petal Lenght of Iris-setosa");
plt.xlabel("Samples");
plt.ylabel("Size");



plt.subplot(1,3,2)
plt.plot(data2["SL"],c='k')
plt.plot(data2["PL"],c='b')
plt.legend(['PL','SL'])
plt.title("Relation between Sepal Lenght and Petal Lenght of Iris-versicolor ");
plt.xlabel("Samples");
plt.ylabel("Size");


plt.subplot(1,3,3)
plt.plot(data3["SL"],c='k')
plt.plot(data3["PL"],c='b')
plt.legend(['PL','SL'])
plt.title("Relation between Sepal Lenght and Petal Lenght of Iris-virginica");
plt.xlabel("Samples");
plt.ylabel("Size");
```

```python
plt.style.use('classic')
plt.plot(data["SW"],c='k')
plt.plot(data["PW"],c='b')
plt.legend(['PW','SW'],loc='lower right')
plt.title("Varition in Sepal width and Petal width");
plt.xlabel("Samples");
plt.ylabel("Size");
```

```python
plt.style.use('seaborn')
plt.bar(data["Classification"],data['SL']);
plt.bar(data["Classification"],data['SW']);
plt.legend(['SL','SW'],loc='upper left');
plt.title("Differance between sepal length and sepal width of different varieties");
plt.xlabel("Classification");
plt.ylabel('Size');
```
